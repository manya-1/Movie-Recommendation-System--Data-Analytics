```python

from sklearn.feature_extraction.text import TfidfVectorizer
import re

# Assuming movie2_features is a DataFrame containing your data

# Concatenate the desired columns into a single column
x = movie2_features['genres'] + ' ' + movie2_features['keywords'] + ' ' + \
    movie2_features['original_language'] + ' ' + movie2_features['overview'] + ' ' + \
    movie2_features['production_companies'] + ' ' + movie2_features['tagline'] + ' ' + \
    movie2_features['title_x'] + ' ' + movie2_features['cast']

# Convert data to lowercase 
def custom_preprocess(text):
    text_lower = text.lower()
   
    return text_lower

# Apply the custom preprocessing to each element in the list
x_processed = [custom_preprocess(text) for text in x]

# Create a TfidfVectorizer instance
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the preprocessed data
tfidf_matrix = tfidf_vectorizer.fit_transform(x_processed)

# Print the TF-IDF matrix (just the shape for brevity)
print("TF-IDF Matrix Shape:", tfidf_matrix.shape)
```

```python
print(x)
```

```python
#get similarity score using cosine similarity 
from sklearn.metrics.pairwise import cosine_similarity
Similarity_score=cosine_similarity(tfidf_matrix)
Similarity_score
```

```python
Similarity_score.shape
```

```python
#get a movie name from user as input and validate for closest spelling
Favourite_movie_name=input('enter ur fav movie name:')
```

```python
all_movielist=movie2_features['title_x'].tolist()
```

```python
import difflib
```

```python
movie_recom=difflib.get_close_matches(Favourite_movie_name,all_movielist)
print(movie_recom)
```

```python
close_match=movie_recom[0]
print(close_match)
```

```python
index_of_close_match=all_movielist.index(close_match)
#we use index function to find the correct index of the close match title
print(index_of_close_match)
```

```python
#geeting a list of similar movies
simi=Similarity_score[index_of_close_match]
Recom_score=list(enumerate(simi))
print(Recom_score)
```

```python
len(Recom_score)
```

```python
#get all movies sort based on recommendation score wrt favourite movie
#sorting the movies baseed on tehir similaritu score
sorted_similar_movies=sorted(Recom_score, key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
#print the name of the movies based on the index
print('top 10 movies suggested for you: \n')
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=movie2_features[movie2_features.index==index]['title_x'].values[0]
    if (i<11):
        print(i,'.', title_from_index)
        i+=1
```

Explanation-In this project, I have developed a movie recommendation system that suggests the top 10 movies based on a user's favorite movie. The recommendation system is built using the TF-IDF (Term Frequency-Inverse Document Frequency) vectorization technique.
